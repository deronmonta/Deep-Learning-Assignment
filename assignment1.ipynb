## Problem 2
We want to show that the Linear regression solution is:
$$w^* = (X^TX)^{-1}X^Ty$$
The equation for Mean Square Error is written as:
$$
MSE = \frac{1}{n} \Sigma _i (\hat y - y)_i^2
$$
Where n is the number of training data and $\hat y$ is obtained by:
$$
\hat y = w^Tx
$$
To obtain the opitmal weight that minimizes the MSE, we can solve for the case where gradient is zero:
$$
\nabla MSE = 0 \\
$$

$$
\nabla \frac{1}{n} ||w^Tx - y||^2_2=0\\
$$

$$
\nabla (w^TX -y)^T(w^TX - y) = 0\\
$$

$$
\nabla (wX^T- y^T)(w^TX - y) = 0\\
$$

$$
\nabla_w (yy^T + w^TX^TX - 2w^TX^T ) = 0\\
$$

$$
2X^TXw-2X^Ty = 0\\
$$

$$
w = (X^TX)^{-1}X^Ty
$$

```python
#Problem 2.3
# 1 Degree of freedom
#x1,y1 with sigma = 0 and N =15
one_col = np.ones((15,1))
x1_1 = np.append(one_col,x1,axis = 1)#Extra column for constant coefficient
w1_1 = np.linalg.inv(x1_1.T.dot(x1_1)).dot(x1_1.T).dot(y1)#Fit the linear regression model 
y1_1_hat = x1_1.dot(w1_1) #This is the fitted value
mse1_1 = np.mean(((y1_1_hat - y1)**2))#Mean square error

# 2 degree of freedom, 
x1_sq = np.square(x1)
x1_2 = np.append(x1, x1_sq,axis = 1)#Add square term to end of array 
x1_2 = np.append(one_col,x1_2,axis = 1)
#Plot the fitted values and ground truth
w1_2 = np.linalg.inv(x1_2.T.dot(x1_2)).dot(x1_2.T).dot(y1)
y1_2_hat = x1_2.dot(w1_2) #This is the fitted value
mse1_2 =  np.mean(((y1_2_hat - y1)**2))

# 9 degrees of freedom
x1_9 = one_col
for i in range(1,10):
    x1_9 = np.append(x1_9,np.power(x1,i),axis = 1)
w1_9 = np.linalg.inv(x1_9.T.dot(x1_9)).dot(x1_9.T).dot(y1)
y1_9_hat = x1_9.dot(w1_9) #This is the fitted value
mse1_9 = np.mean((y1_9_hat - y1)**2)
plt.plot(x1,y1,'ro')#Plot ground truth value
plt.plot(x1_1[:,1],y1_1_hat,'gx')#plot the predicted value for dof = 1
plt.plot(x1_2[:,1],y1_2_hat,'bx')#plot the predicted value for dof = 2
plt.plot(x1_9[:,1],y1_9_hat,'y.')#plot the predicted value for dof = 9
plt.xlabel('Degree of freedom = 1,2,9, sigma = 0 and N =15')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors as NN
data = np.loadtxt('data_seed.dat',dtype = None)
#print(data.shape)#Inspect the size of data
data = data.reshape(210,8)


#Here, we want to normalize the data
for i in range(0,7):
    data[:,i] = (data[:,i] - np.mean(data[:,i])) / np.std(data[:,i])
label = data[:,-1]
np.random.shuffle(data)
#This is for 5 fold croos validation
five_fold = np.split(data,5)

error1 = 0
error5 = 0
error10 = 0
error15 = 0
#For leave one out cross validation
for i in range(0,len(data)):
    cv = data
    single = data[i:i+1,:]#Single test point to leave out 
    cv = np.delete(data, i, axis=0)#Remove single point from training data

    #For 1 neighborhood
    nbrs1 = NN(n_neighbors=1, algorithm='auto').fit(cv[:,:7])
    dist,indices = nbrs1.kneighbors(single[:,:7])#Query 
    votes = data[indices,7]
   
    (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
    prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
    
    if prediction != single[:,-1]:  #Cross Validation error as wrongly classified data
        error1 +=1
        
        
    #For 5 neighborhood
    
    nbrs5 = NN(n_neighbors=5, algorithm='auto').fit(cv[:,:7])
    dist,indices = nbrs5.kneighbors(single[:,:7])
    votes = data[indices,7]
    (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
    prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
    
    if prediction != single[:,-1]:
        error5 +=1 #Report test error
    
        
    #For 10 neighborhood
    
    nbrs10 = NN(n_neighbors=10, algorithm='auto').fit(cv[:,:7])
    dist,indices = nbrs10.kneighbors(single[:,:7])
    votes = data[indices,7]
    (values,counts) = np.unique(votes,return_counts=True) #Counting the votes, return label with most votes
    prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
    
    if prediction != single[:,-1]:
        error10 +=1
        
        
    #For 15 neighborhood
    nbrs15 = NN(n_neighbors=15, algorithm='auto').fit(cv[:,:7])
    dist,indices = nbrs15.kneighbors(single[:,:7])
    votes = data[indices,7]
    (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
    prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
    if prediction != single[:,-1]:
        error15 +=1 
        
        
print(error1)
print(error5)
print(error10)
print(error15)
x = [1,5,10,15]
test_error = [error1,error5,error10,error15]
plt.plot(x,test_error,'ro')#Plot ground truth value
plt.xlabel('Test error for leave one out ')
plt.show()
error1 = []
error5 = []
error10 = []
error15 = []


#For five fold cross validation
for i in range(0,5):
    train = five_fold
    test = five_fold[i]
    train = np.delete(five_fold,i, axis=0)#Remove cross validation points from training data 
    train = train.reshape(168,8)
    
    #One neighbor
    nbrs1 = NN(n_neighbors=1, algorithm='auto').fit(train[:,:7])
    errors = 0
    for point in test:
        distances, indices = nbrs1.kneighbors(point[:7].reshape(1,-1))
        votes = train[indices,7]
        
        (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
        prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
        if prediction != point[7]:
            errors +=1
    error1.append(errors)
    
    #5 neighbor
    nbrs5 = NN(n_neighbors=5, algorithm='auto').fit(train[:,:7])
    errors = 0
    for point in test:
        distances, indices = nbrs5.kneighbors(point[:7].reshape(1,-1))
        votes = train[indices,7]
        (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
        prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
        if prediction != point[7]:
             errors +=1
    error5.append(errors)
            
    #10 neighbor
    nbrs10 = NN(n_neighbors=10, algorithm='auto').fit(train[:,:7])
    errors = 0
    for point in test:
        distances, indices = nbrs10.kneighbors(point[:7].reshape(1,-1))

        votes = train[indices,7]
        (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
        prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
        if prediction != point[7]:
             errors +=1
    error10.append(errors)
            
    #15 neighbors
    nbrs15 = NN(n_neighbors=15, algorithm='auto').fit(train[:,:7])
    errors = 0
    for point in test:
        distances, indices = nbrs15.kneighbors(point[:7].reshape(1,-1))
        votes = train[indices,7]
        (values,counts) = np.unique(votes,return_counts=True)#Counting the votes, return label with most votes
        prediction = values[np.argmax(counts)] #Prediction is the label with the most votes
        if prediction != point[7]:
             errors +=1
    error15.append(errors)
    
         #Report test error
print(sum(error1))
print(sum(error5))
print(sum(error10))
print(sum(error15))
x = [1,5,10,15]
test_error = [np.mean(error1),np.mean(error5),np.mean(error10),np.mean(error15)]
plt.plot(x,test_error,'ro')#Plot ground truth value
plt.xlabel('Test error for 5 fold cross validation')
plt.show()
```
